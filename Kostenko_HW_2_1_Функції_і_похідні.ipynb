### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def f(x):
  return (10*x-5)
```

```python
xs = np.linspace(0,10,20)
f = f(xs)
d = np.full_like(xs, 10)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(xs, f, label='f(x) = 10x - 5')
plt.plot(xs, d, label="f'(x) = 10", linestyle='--')

plt.xlabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è')
plt.ylabel('–û—Ü—ñ–Ω–∫–∞ / –ü–æ—Ö—ñ–¥–Ω–∞')
plt.title('–§—É–Ω–∫—Ü—ñ—è –æ—Ü—ñ–Ω–∫–∏ —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞')
plt.legend()
plt.grid(True)
plt.show()
```

**–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó f(x) - –ø—Ä—è–º–∞ –ª—ñ–Ω—ñ—è**

–ü—Ä–∏ –∑–±—ñ–ª—å—à–µ–Ω–Ω—ñ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è –æ—Ü—ñ–Ω–∫–∞ –∑—Ä–æ—Å—Ç–∞—î –ª—ñ–Ω—ñ–π–Ω–æ

–ö–æ–∂–Ω–∞ –¥–æ–¥–∞—Ç–∫–æ–≤–∞ –≥–æ–¥–∏–Ω–∞ –¥–∞—î –æ–¥–Ω–∞–∫–æ–≤–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç

**–ì—Ä–∞—Ñ—ñ–∫ –ø–æ—Ö—ñ–¥–Ω–æ—ó f‚Ä≤(x)**

–ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞ –ª—ñ–Ω—ñ—è –Ω–∞ —Ä—ñ–≤–Ω—ñ 10

–¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ —à–≤–∏–¥–∫—ñ—Å—Ç—å –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω–∫–∏ –Ω–µ –∑–º—ñ–Ω—é—î—Ç—å—Å—è

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
x = sp.symbols('x')
```

```python
#–°–∏–Ω—É—Å–æ—ó–¥–Ω–∞
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

#–ª–æ–≥–∞—Ä–∏—Ñ–º—ñ—á–Ω–∞
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)


print("–ü–æ—Ö—ñ–¥–Ω–∞ —Å–∏–Ω—É—Å–æ—ó–¥–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó sin(ùë•):", derivative_f_sin)
print("–ü–æ—Ö—ñ–¥–Ω–∞ –ª–æ–≥–∞—Ä–∏—Ñ–º—ñ—á–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó ln(ùë•):", derivative_f_ln)
```

```python
f_sin_num = sp.lambdify(x, f_sin, 'numpy')
derivative_num = sp.lambdify(x, derivative_f_sin, 'numpy')
```

```python
xs = np.linspace(0,10,200)
```

```python
ys = f_sin_num(xs)
dys = derivative_num(xs)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(xs, ys, label='f(x) = sin(x)')
plt.plot(xs, dys, label="f'(x) = cos(x)", linestyle='--')
plt.legend();
```

```python
f_ln_num =  sp.lambdify(x, f_ln, 'numpy')
derivative_ln_num = sp.lambdify(x, derivative_f_ln, 'numpy')
```

```python
fs = f_ln_num(xs)
dfs = derivative_ln_num(xs)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(xs, fs, label='f(x) = ln(x)')
plt.plot(xs, dfs, label="f'(x) = 1/x", linestyle='--')
plt.legend();
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
f_comp = x**2 - 5 * x + 1
derivative_f_comp = sp.diff(f_comp, x)
```

```python
xs = np.linspace(0,100,200)
```

```python
f_comp_num = sp.lambdify(x, f_comp, 'numpy')
derivative_f_comp_num = sp.lambdify(x, derivative_f_comp, 'numpy')
```

```python
ys = f_comp_num(xs)
dys = derivative_f_comp_num(xs)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(xs, ys, label='f(x)=x^2‚àí5x+1')
plt.plot(xs, dys, label="f'(x) = 2x + 5", linestyle='--')

plt.xlabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å (x)')
plt.ylabel('–í–∏—Ç—Ä–∞—Ç–∏ / –ü–æ—Ö—ñ–¥–Ω–∞ –≤–∏—Ç—Ä–∞—Ç')
plt.title('–§—É–Ω–∫—Ü—ñ—è –≤–∏—Ç—Ä–∞—Ç —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞')
plt.legend()
plt.grid(True)
plt.show();
```
